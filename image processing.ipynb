```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("Screenshot 2026-04-07 at 2.23.34 PM.png") # this changes, need you image name
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale (intensity = brightness)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Normalize brightness (0 to 1)
gray_norm = gray / 255.0

# Compute mixing index
mean_intensity = np.mean(gray)
std_intensity = np.std(gray)

mixing_index = 1 - (std_intensity / mean_intensity)

print("Mean Intensity:", mean_intensity)
print("Std Intensity:", std_intensity)
print("Mixing Index:", mixing_index)

# Show grayscale image
plt.imshow(gray, cmap='gray')
plt.show()
```
